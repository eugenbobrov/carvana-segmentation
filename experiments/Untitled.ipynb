```python
import os
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
```

```python
segnet_loss = np.load('history_segnet_gray.npy').tolist()['loss']
segnet_val_loss = np.load('history_segnet_gray.npy').tolist()['val_loss']
unet_loss = np.load('history_unet_gray.npy').tolist()['loss']
unet_val_loss = np.load('history_unet_gray.npy').tolist()['val_loss']
unetdrop_loss = np.load('history_unetdrop_gray.npy').tolist()['loss']
unetdrop_val_loss = np.load('history_unetdrop_gray.npy').tolist()['val_loss']

plt.style.use('ggplot'); plt.figure()
plt.plot(unet_loss); plt.plot(segnet_loss); plt.plot(unetdrop_loss)
plt.title('model train loss'); plt.legend(['unet', 'segnet', 'unetdrop'])
plt.ylabel('loss'); plt.xlabel('epoch'); plt.yscale('log')
plt.savefig('architecture_train_loss.png'); plt.show()

plt.style.use('ggplot'); plt.figure()
plt.plot(unet_val_loss); plt.plot(segnet_val_loss); plt.plot(unetdrop_val_loss)
plt.title('model valid loss'); plt.legend(['unet', 'segnet', 'unetdrop'])
plt.ylabel('loss'); plt.xlabel('epoch'); plt.yscale('log')
plt.savefig('architecture_valid_loss.png'); plt.show()
```

```python
sgd_loss = np.load('history_sgd.npy').tolist()['loss']
sgd_val_loss = np.load('history_sgd.npy').tolist()['val_loss']
adam_loss = np.load('history_gray_adam_unet.npy').tolist()['loss']
adam_val_loss = np.load('history_gray_adam_unet.npy').tolist()['val_loss']

plt.style.use('ggplot'); plt.figure()
plt.plot(sgd_loss); plt.plot(adam_loss)
plt.title('model train loss'); plt.legend(['simple sgd', 'adam'])
plt.ylabel('loss'); plt.xlabel('epoch'); plt.yscale('log')
plt.savefig('optimizer_train_loss.png'); plt.show()

plt.style.use('ggplot'); plt.figure()
plt.plot(sgd_val_loss); plt.plot(adam_val_loss)
plt.title('model valid loss'); plt.legend(['simple sgd', 'adam'])
plt.ylabel('loss'); plt.xlabel('epoch'); plt.yscale('log')
plt.savefig('optimizer_valid_loss.png'); plt.show()
```

```python
color_loss = np.load('history_unet_color.npy').tolist()['loss']
color_val_loss = np.load('history_unet_color.npy').tolist()['val_loss']
gray_loss = np.load('history_unet_gray.npy').tolist()['loss']
gray_val_loss = np.load('history_unet_gray.npy').tolist()['val_loss']

plt.style.use('ggplot'); plt.figure()
plt.plot(color_loss); plt.plot(gray_loss)
plt.title('model train loss'); plt.legend(['color', 'gray'])
plt.ylabel('loss'); plt.xlabel('epoch'); plt.yscale('log')
plt.savefig('color_train_loss.png'); plt.show()

plt.style.use('ggplot'); plt.figure()
plt.plot(color_val_loss); plt.plot(gray_val_loss)
plt.title('model valid loss'); plt.legend(['color', 'gray'])
plt.ylabel('loss'); plt.xlabel('epoch'); plt.yscale('log')
plt.savefig('color_valid_loss.png'); plt.show()
```

```python
loss = np.load('history_gray_adam_unet.npy').tolist()['loss']
val_loss = np.load('history_gray_adam_unet.npy').tolist()['val_loss']

accuracy = np.load('history_gray_adam_unet.npy').tolist()['acc']
val_accuracy = np.load('history_gray_adam_unet.npy').tolist()['val_acc']

plt.style.use('ggplot'); plt.figure()
plt.plot(loss); plt.plot(val_loss)
plt.title('model loss'); plt.legend(['train loss', 'validation loss'])
plt.ylabel('loss'); plt.xlabel('epoch'); plt.yscale('log')
plt.savefig('final_loss.png'); plt.show()

plt.style.use('ggplot'); plt.figure()
plt.plot(accuracy); plt.plot(val_accuracy)
plt.title('model accuracy'); plt.legend(['train accuracy', 'validation accuracy'])
plt.ylabel('loss'); plt.xlabel('epoch'); plt.yscale('log')
plt.savefig('final_accuracy.png'); plt.show()
```
